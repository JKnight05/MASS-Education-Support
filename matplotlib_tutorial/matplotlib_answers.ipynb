## Exercise 3 - Subplots with mean and Deviation

```python
mean_temp = np.mean(list(results.values()), axis=0)
deviation = temperature - mean_temp

fig, axes = plt.subplots(2, 1, figsize=(6, 6), sharex=True)

axes[0].plot(time, mean_temp, linewidth=2)
axes[0].set_ylabel("Mean Temperature (°C)")
axes[0].set_title("Average Temperature Over Time")
axes[0].grid(True)

axes[1].plot(time, deviation)
axes[1].set_xlabel("Time (minutes)")
axes[1].set_ylabel("Deviation (°C)")
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

## Exercise 4 -- highlight and Annotate Max

```python
idx = np.argmax(mean_temp)

fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(time, mean_temp, label="Mean Temperature")
ax.scatter(time[idx], mean_temp[idx], zorder=3)

ax.annotate(
    f"Max = {mean_temp[idx]:.2f} °C",
    xy=(time[idx], mean_temp[idx]),
    xytext=(time[idx] + 1, mean_temp[idx] + 0.5),
    arrowprops=dict(arrowstyle="->")
)

ax.set_xlabel("Time (minutes)")
ax.set_ylabel("Temperature (°C)")
ax.set_title("Maximum Average Temperature")
ax.legend()
ax.grid(True)

plt.show()
```
