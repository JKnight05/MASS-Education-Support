```python
trials = 5
results = {
    f"Trial {i+1}": temperature + np.random.normal(0, 0.5, size=time.size)
    for i in range(trials)
}
```

### Side Note
What makes the difference between a good v bad plot?

A bad plot:

```python
plt.plot(time, temperature)
plt.plot(time, temperature + np.random.normal(0, 1, size=len(time)))
plt.plot(time, temperature - np.random.normal(0, 1, size=len(time)))
plt.show()
```

What's wrong? 

- No labels 
- No title 
- No Legend 
- Meaning unclear 
- Multiple lines with no explanation

A good plot (same data, better choices)

```python
fig, ax = plt.subplots(figsize=(7, 4))

for name, data in results.items():
    ax.plot(time, data, alpha=0.6, label=name)

ax.set_title("Temperature Measurements Across Trials")
ax.set_xlabel("Time (minutes)")
ax.set_ylabel("Temperature (°C)")
ax.legend(ncol=2)
ax.grid(True)

plt.tight_layout()
plt.show()
```

Better because: 

- Clear meaning 
- Proper labels 
- Transparency (`alpha`) reduces clutter

## Exercise 5 Publication-Style Figure




```python
laps = np.arange(1, 26)

# Fuel load (kg)
fuel_start = 105
fuel_burn_per_lap = 1.8
fuel = fuel_start - fuel_burn_per_lap * laps

# Tire wear (0 = new, 1 = fully worn)
tire_wear = 1 - np.exp(-laps / 18)

# Lap time (seconds)
base_lap_time = 92.0
lap_time = (
    base_lap_time
    + 0.25 * tire_wear * 10          # tire degradation
    - 0.03 * (fuel_start - fuel)     # lighter fuel is faster
    + np.random.normal(0, 0.15, size=len(laps))
)

# Top speed (km/h)
top_speed = (
    330
    - 8 * tire_wear
    + 0.05 * (fuel_start - fuel)
    + np.random.normal(0, 1.0, size=len(laps))
)

# ERS deployment (MJ per lap)
ers = (
    4
    - 0.02 * laps
    + np.random.normal(0, 0.1, size=len(laps))
)
ers = np.clip(ers, 2.5, 4.0)

# Lap time delta relative to fastest lap
delta_lap_time = lap_time - lap_time.min()

# Simple performance index (arbitrary)
performance_index = (
    0.6 * (top_speed / top_speed.max())
    + 0.4 * (1 - tire_wear)
)

print("Fastest lap:", lap_time.min())
print("Slowest lap:", lap_time.max())
print("Final tire wear:", tire_wear[-1])
print("Final fuel load:", fuel[-1])
```
